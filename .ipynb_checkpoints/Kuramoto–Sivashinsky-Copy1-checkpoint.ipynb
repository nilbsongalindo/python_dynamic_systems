## Kuramoto-Sivashinky Equation

Is a fourth-order nonlinear partial differential equation, which gives us a model for the temporal evolution of the combustion wavefront in a convection flame. One particularity of the KS equation, is that possess chaotic behavior and it's hard to integrate by traditional methods, like runge-kutta

![KS](data/KS.png)

```python
from numpy import *
#import numpy as np
from random import random
from scipy.integrate import odeint
from matplotlib.pyplot import *
from numpy.fft import rfft, irfft
```

The real Kuramoto-Sivashinsky eq., according to the internet:
yt = -y4x -yxx -0.5*abs(yx)**2  (1) 

(On Flame Propagation Under Conditions of Stoichiometry, SIAM Journal on Applied Mathematics, 1980, Vol. 39, No. 1 : pp. 67-82). 



An alternate form of KS eq. modified by the addition of a spatial inhomogeneity (source term):


![KS_alternate](data/alternate.png)

 (Model-Free Prediction of Large Spatiotemporally Chaotic Systems from Data, Phys. Rev. Lett. (2018), 120, 024101). Cf. also Fluctuation of mean Lyapunov exponent for Kuramoto-Sivashinksky equation, Physica A (2000), 285, 325-331.)


```python
L = 200.0
n_lamb = 2 # number of wavelengths characteristic of the equation.
Q = 512 #number of points in space discretization. L will be divided into Q points.
lamb = L/n_lamb
mu = 0.01 # amplitude of inhomogeneity.
tf = 500.0 # final simulation time.
dt = 0.01 # step-size.
Nt = int(tf/dt)+1 # number of points in time.
Nx = Q
dx = L/(Nx+1) # separation size between points in space;

Omega_0 = 2*pi/L
Omega_inh = 2*pi/lamb
index_inh = n_lamb # index to identify the number of oscillations.

#spatial frequencies (appearing in Fourier analysis of real signals):
rfreqs = linspace(0, Omega_0*Nx/2, int(Nx/2)+1)
# Fourier multipliers forming the linear operator
LinOp = rfreqs**2 - rfreqs**4
LinOp[index_inh] += mu 
t = linspace(0.0, tf, Nt)
x = linspace(0.0, L-dx, Nx)

def ddx(y):
    ### forward difference
    #return (y-roll(y,1))/dx
    ### backward difference
    #return (roll(y,-1)-y)/dx
    ### centered difference
    return (roll(y,-1)-roll(y,1))/(2*dx)
```

```python
### initial condition (please use a period-L function):
y0 = cos(2*pi*x/L)*(1+sin(2*pi*x/L))
Y0 = rfft(y0)

y = zeros((Nx, Nt))
y[:, 0] = y0 
Yt = Y0.copy()

for i in range(Nt-1):
    IntFacL = exp(dt*LinOp)
    yx = ddx(y[:,i])
    
    NLop = -yx #*y[:,i]    
    IntFacNL = exp(dt*NLop)
    y[:,i+1] = irfft(IntFacL*rfft(IntFacNL*y[:,i]))
```

```python
figure(1, figsize = (8,4.5))
subplot(211)
imshow(y, cmap = cm.jet, extent = [0,tf, 0, L-dx], aspect = 'auto')
subplot(212)
plot(x,y[:,0]) #initial condition
plot(x,y[:,-1]) #spatial solution at the last value of time
```
